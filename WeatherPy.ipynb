## Generate Cities List

```python
# List for holding lat_lngs and cities
lat_lngs = []
cities = []

# Create a set of random lat and lng combinations
lats = np.random.uniform(low=-90.000, high=90.000, size=1500)
lngs = np.random.uniform(low=-180.000, high=180.000, size=1500)
lat_lngs = zip(lats, lngs)

# Identify nearest city for each lat, lng combination
for lat_lng in lat_lngs:
    city = citipy.nearest_city(lat_lng[0], lat_lng[1]).city_name
    
    # If the city is unique, then add it to a our cities list
    if city not in cities:
        cities.append(city)

# Print the city count to confirm sufficient count
len(cities)
```

### Perform API Calls
* Perform a weather check on each city using a series of successive API calls.
* Include a print log of each city as it'sbeing processed (with the city number and city name).


```python
url = "http://api.openweathermap.org/data/2.5/weather?"
units = "imperial"

# Set up lists to hold responses
name = []
lat = []
temp = []
humidity = []
clouds = []
wind = []

for city in cities:
    query_url = url + "appid=" + api_key + "&q=" + city + "&units=" + units
    weather_response = requests.get(query_url)
    weather_json = weather_response.json()
    
    print(f"Retrieving Results for {city} ...")
    
    try:
        print(f"Results Found for {city}.")
        
        name.append(weather_json["name"])
        lat.append(weather_json["coord"]["lat"])
        temp.append(weather_json["main"]["temp"])
        humidity.append(weather_json["main"]["humidity"])
        clouds.append(weather_json["clouds"]["all"])
        wind.append(weather_json["wind"]["speed"])
        
    except KeyError:
        print("Missing field/result... skipping")
    
    print("--------------------")
        
```

```python
# Confirm we have enough records

weather_data = pd.DataFrame({"City": name, 
                            "Lattitude": lat, 
                            "Temperature": temp, 
                            "Humidity": humidity,
                            "clouds": clouds,
                            "wind": wind})

weather_data.count()
```

#### Latitude vs. Temperature Plot

```python
# Create scatter plots for each city type (this method allows us to create proper legends)

plt.scatter(lat, temp,
            s=20, c="red", edgecolor="black", 
            linewidths=0.5, alpha=0.5, marker="o")

# Incorporate the other graph properties
# Add titles and labels
plt.grid(True)
plt.title("City Latitude vs. Max Temperature")
plt.xlabel("Latitude")
plt.ylabel("Temerature(F)")

# Save file to output folder
plt.savefig("./output/lat_v_temp.png")

# Display plot
plt.show()
```
